Supervised Learning :

Supervised Learning is a type of machine learning where the data comes with labels or correct answers, and the model learns to map inputs to outputs.


Unsupervised Learning :

Unsupervised Learning is when the data has no labels, and the model tries to discover hidden patterns, groups, or relationships in the data.
    

Reinforcement Learning :
    
Reinforcement Learning is when an agent learns by interacting with an environment and receives rewards or penalties based on its actions.

Comparison of Machine Learning Types :

| ML Type | Description | Real-World Example 1 | Real-World Example 2 |
|--------|------------|----------------------|----------------------|
| Supervised Learning | Learning from labeled data where correct answers are known | Spam Detection | House Price Prediction |
| Unsupervised Learning | Finding hidden patterns or clusters in unlabeled data | Customer Segmentation | Fraud/Anomaly Detection |
| Reinforcement Learning | Learning through rewards and penalties by interacting with environment | Self-Driving Cars | Game Playing AI |
